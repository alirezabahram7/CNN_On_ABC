## Heteroscedastic Loss

### Define train loss between network predicted output and desired posteriors.

```python
def het_loss(actual_params, pred_output):
    vector_size = pred_output.shape[1] // 2
    posterior_expectation = pred_output[:,:vector_size]
    variance = pred_output[:,vector_size:]
    exp_diff = (actual_params - posterior_expectation)
    het_loss = keras.backend.sum((1 / variance) * (exp_diff**2.) + (keras.backend.log(variance)), -1)
    return het_loss
```

# DeepInference

### The model (network) is constructed in this function using given arguments.
### The input vector is processed using one dimensional convolutional filters (multiple filters with respect to number of channels).
### It is processed through global average pooling layers and one fullly connected layer is attached at the end.
### Finally the model produced the network output.

```python
def deep_inference(param_num, channels_num, filter_sizes_array):
    
    inp = keras.Input(shape=(None, channels_num))
    x = keras.layers.Conv1D(filter_sizes_array[0], kernel_size=1, strides=1, activation='relu')(inp)
    
    for filter_size in filter_sizes_array[1:]:
        x = keras.layers.Conv1D(filter_size, kernel_size=3, strides=2, activation='relu')(x)
    
    x = keras.layers.GlobalAveragePooling1D()(x)
    mean = keras.layers.Dense(param_num)(x)
    variance = keras.layers.Dense(param_num, activation='softplus')(x)
    
    out = keras.layers.Concatenate()([mean, variance])
    
    model = keras.Model(inp, out)
    
    return model
```

# Plot

### Plotting and quantitative analysis.
### Plot various network outputs against desired values.
### Plot function gives predicted and desired values and data names, such as $\mathrm{E}[\mu_2 | X]$ or $Var[\mu_1 | X]$

### After plotting it calculates RMSE and correlation coefficient for quantify network predictions.

```python
def plot(true_values, predicted_values, feature_names):

    normalized_maes = []

    figure, axes = plt.subplots(2, true_values.shape[1] // 2, figsize=(15, 10))

    three_places = Decimal(10) ** -3 

    for i, ax in enumerate(axes.flat):
        ax.plot(true_values[:, i], predicted_values[:, i], marker='o', linewidth=0, alpha=0.2, color='green')
        
        if i == 0:
            ax.set_xlabel('True')
            ax.set_ylabel('Estimated')
        ax.set_title(feature_names[i])
        
        correlation = np.corrcoef(true_values[:, i], predicted_values[:, i])[0, 1]
        correlation_display = str(Decimal(correlation).quantize(three_places))

        mae = np.round(np.mean(abs((true_values[:, i] - predicted_values[:, i]) / true_values[:, i])), 2)
        rmse = np.sqrt(np.mean((true_values[:, i] - predicted_values[:, i]) ** 2))
        rmse_display = str(Decimal(rmse).quantize(three_places))
        normalized_maes.append(mae)

        ax.text(0.2, 0.9, f'RMSE: {rmse_display}', ha='center', va='center', transform=ax.transAxes, color="brown")
        ax.text(0.15, 0.85, f'r: {correlation_display}', ha='center', va='center', transform=ax.transAxes, color="purple")

        #for variances
        if i >= true_values.shape[1] // 2:
            ax.set_xscale('log')
            ax.set_yscale('log')
    
        x = np.array(ax.get_xlim())
        #plot y = x
        ax.plot(x, x, color='blue')

        ax.set_xlim(np.min(true_values[:, i]), np.max(true_values[:, i]))
        ax.set_ylim(np.min(true_values[:, i]), np.max(true_values[:, i]))

    plt.show()
    return None
```

## Toy Example1

```python
PARAMETER_NAMES1 = [r'$\mathrm{E}[\mu_1 | X]$',
         r'$\mathrm{E}[\mu_2 | X]$',
         r'$\mathrm{E}[\sigma_1 | X]$',
         r'$\mathrm{E}[\sigma_2 | X]$',
         r'$Var[\mu_1 | X]$',
         r'$Var[\mu_2 | X]$',
         r'$Var[\sigma_1 | X]$',
         r'$Var[\sigma_2 | X]$']
plot(y_test1, preds1,PARAMETER_NAMES1)
```
